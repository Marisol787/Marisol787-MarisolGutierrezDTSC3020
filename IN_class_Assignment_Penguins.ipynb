## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='deep'
)
plt.title("Bill Length vs. Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

**Your interpretation (2–3 sentences):**

_The three species all seem to get concentrated within their groups. It seeems like there is no linearity between any of the species.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=10, color='skyblue')
plt.title("Histogram of Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=30, color='salmon')
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()
```

**Your comparison (1–2 sentences):**

THe first histogram seems to follow a normal bellcurve. THe second one has a dip in the middle creating a bimodal curve.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='pastel'
)
plt.title("Body Mass by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

**Your summary (1–2 sentences):**

THe body mass of the first species doesnt have any outliers in conrtast to the second species which shows one outlier. The third species follows a normal whisker plot.
